Basic Data Analysis

```python
import pandas as pd

# Creating a sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
}

df = pd.DataFrame(data)

# Basic analysis
print(df.describe())  # Summary statistics
print(df.info())      # Data types and missing values
print(df['Age'].mean())  # Mean of Age column
```

Filtering Data in a DataFrame

```python
import pandas as pd

# Creating a sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
}

df = pd.DataFrame(data)

# Filtering employees with Salary > 60000
filtered_df = df[df['Salary'] > 60000]

print(filtered_df)
```

Titanic Dataset Analysis in Google Colab

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic dataset from seaborn
df = sns.load_dataset('titanic')

# Display first few rows
print(df.head())

# Basic exploration
print(df.info())
print(df.describe())

# Countplot of survived passengers
sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.show()

# Boxplot of age distribution by survival
sns.boxplot(x='survived', y='age', data=df)
plt.title("Age Distribution by Survival")
plt.show()
```
